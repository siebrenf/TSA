## make the above into a function

```python
import json
import subprocess as sp

def query_orthodb(query):
    """
    see https://www.orthodb.org/orthodb_userguide.html#api
    """
    url = f"https://www.orthodb.org/{query}"
    response = sp.check_output(f"wget -O - '{url}'", shell=True)
    data = json.loads(response)["data"]
    return data


def get_othoclusters(query, level=None, species=None):
    """
    ask orthoDB which ortholog clusters the gene is found in 
    
    query: a gene name/uniprot ID
    level: NCBI taxon id of the clade
    species: list of NCBI numerical taxonomy ids*
    
    *this option is not documented on OrthoDB
    
    returns a list of cluster IDs
    """
    query = f"search?query={query}"
    if level:
        query += f"&level={level}"
    if isinstance(species, str):
        query += f"&species={species}"
    elif isinstance(species, list):
        query += f"&species={','.join(species)}"
    return query_orthodb(query)


def get_orthologs(cluster_id, species=None):
    """
    get orthologous genes for each specie in one othocluster
    
    returns a dict with species, genes and metadata
    """
    query = f"orthologs?id={cluster_id}"
    if isinstance(species, str):
        query += f"&species={species}"
    elif isinstance(species, list):
        query += f"&species={','.join(species)}"
    return query_orthodb(query)


def x2y(gene, level, species, names, strict=True):
    """
    for the given gene, return a list of orthologous genes, using OrthoDB.
    
    strict: if set to True (default), only return genes of the 
    target gene. If False, also return othrologs of near hits.
    """
#     level = 7742
#     species = ["9606_0", "7955_0"]  # human and zebrafish
    cluster_ids = get_othoclusters(gene, level=level, species=species)
    
    # list all genes per cluster per specie
    clusters = {}
    for c in cluster_ids:
        clusters[c] = {}
        
        # each member of the ortho group is a specie dict with name, genes & more
        ortho_grp = get_orthologs(c, species=species)
        for specie in ortho_grp:
            name = specie['organism']['name']
            genes = [g["gene_id"]['id'] for g in specie['genes']]
            clusters[c][name] = genes
            
        if not all(n in clusters[c] for n in names):  # 'Homo sapiens' not in clusters[c] or 'Danio rerio' not in clusters[c]:
            del clusters[c]
    
    # multiple clusters may contain references to our gene but not the actual gene
    # if strickt is True, filter out clusters where the human genes lack the target gene
    if strict is True:
        ok_clusters = {}
        for c in clusters:
            if gene in clusters[c][names[0]]:
                ok_clusters[c] = clusters[c]
        clusters = ok_clusters
    
    # list the orthologs
    orthologs = []
    for c in clusters:
        orthologs.extend(clusters[c][names[1]])
        
    # get unique genes
    orthologs = list(set(orthologs))
    return orthologs


gene = 'nmnat2'
level = 7742
species = ["7955_0", "8364_0"]
names = ["Danio rerio", "Xenopus tropicalis"]
x2y(gene, level, species, names, strict=False)
```

```python
mydict = {}
for n, gene in enumerate(overlap):
    print(round(100* n/len(overlap), 2), "%")
    ortholog = x2y(gene, level, species, names, strict=True)
    mydict[gene] = ortholog
```
